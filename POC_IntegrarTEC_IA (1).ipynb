Esta Proof of Concept (POC) tiene como objetivo demostrar cómo la aplicación de técnicas de Fast Prompting puede mejorar la generación de contenidos de orientación laboral para jóvenes egresados de programas de formación tecnológica como IntegrarTEC.

## Problema

Los jóvenes egresados de programas de formación tecnológica enfrentan dificultades para traducir su formación en oportunidades laborales concretas. Esta POC busca demostrar cómo la IA generativa puede asistir en la construcción de CVs optimizados mediante el uso estratégico de prompts.

## Objetivos

### Objetivo general

Acompañar la inserción en el mercado laboral tecnológico de jóvenes egresados del programa IntegrarTEC, mediante el desarrollo de una herramienta basada en inteligencia artificial generativa que facilite la construcción de perfiles laborales y reduzca barreras de acceso al empleo.

### Objetivos específicos

- Contribuir al acceso a empleos de calidad en el sector tecnológico para jóvenes provenientes de contextos vulnerables.
- Fortalecer las estrategias de orientación laboral mediante herramientas escalables y accesibles.
- Facilitar la traducción de experiencias informales y educativas en competencias laborales relevantes.
- Reducir la brecha entre la formación técnica adquirida y las exigencias del mercado IT.
- Proponer un modelo de acompañamiento automatizado que complemente el trabajo de orientadores humanos.

```python
## Metodología

El desarrollo de la POC se llevó a cabo siguiendo los siguientes pasos:

1. Definición del problema: Se identificó la dificultad que enfrentan jóvenes egresados de programas tecnológicos al momento de construir un CV orientado al mercado IT.

2. Selección de caso de prueba: Se construyó un perfil basado en características típicas de egresados de IntegrarTEC.

3. Diseño de prompt base: Se elaboró un prompt general para generar un CV.

4. Aplicación de técnicas de Fast Prompting:
   - Role Prompting: Se asignó al modelo el rol de orientador laboral especializado en perfiles junior IT.
   - Structured Prompting: Se definieron secciones específicas para organizar el output.
   - Instrucciones claras y delimitadas.

5. Generación y comparación de outputs: Se analizaron los resultados obtenidos para evaluar mejoras en claridad, estructura y orientación profesional.

6. Análisis crítico: Se evaluó cómo la optimización del prompt impacta en la calidad y pertinencia del resultado.
```

```python
perfil = """
Nombre: Pedro B.
Edad: 22
Programa: IntegrarTEC - Cohorte 1
Estado: Egresado
Actualmente trabaja en comercio informal.
No posee experiencia profesional en IT.
Realizó proyectos académicos en programación web.
"""
```

## Aplicación de Técnicas de Fast Prompting

En el segundo prompt se aplicaron las siguientes técnicas:

- Role Prompting: Se asignó un rol específico al modelo (orientador laboral IT).
- Structured Prompting: Se definieron secciones concretas para mejorar la claridad del output.
- Instrucciones específicas: Se establecieron límites y formato deseado.

Estas técnicas permiten obtener resultados más consistentes y alineados con el objetivo del proyecto. El prompt optimizado mejora la precisión, coherencia y orientación laboral del resultado, demostrando la importancia del diseño estratégico de prompts.
